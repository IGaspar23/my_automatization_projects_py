## In this data cleaning process we will do:
1. Delete redundant columns
2. Rename columns
3. Dropping duplicates 
4. Remove the NaN values
5. Check for some more transformations


```python
# First, we notice that the "Unnamed" column is redundant because it gives a integer number for each row.
df = df.drop(columns= "Unnamed: 0")
```

```python
df.head()
```

```python
# For renaming the colums we execute the following chunk of code 
df_v2= df.rename(columns={"Deatils" : "Details"})
df_v2.head()
```

```python
# 3. In the case for duplicates, we look for them by
df_v2.duplicated().value_counts().plot(kind="bar")
```

```python
# 3. Here we observe 2827 duplicated rows in our dataset, for dropping them we use 
df.drop_duplicates(inplace=True)
df.duplicated().value_counts()
df.duplicated().sum()
df
```

```python
#We do not observe duplicates, but we notice a discrepancy in the number of rows. They are unsorted!!. To fix this we write
df.reset_index(inplace=True) # The inplace within the brackets reset our dataframe with the changes we are performing
df
```

```python
# For dropping duplicates we execute  -> df.drop(columns={"Unnamed: 0", "index"})
df
```

```python
# 4. To check if the dataset has nan values we type
df.isna()
```

```python
# As we saw, there are not nan values. Just for to be sure about it, we press
df. dropna(inplace=True)
```

```python
# We corroborate 
df.isna().sum()
```
